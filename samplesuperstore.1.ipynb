```python
g= sns.FacetGrid(df,row="Region",col="Segment",hue="Ship Mode",palette="Set1")
kws = dict(s=50,linewidth=0.5,edgecolor="blue")
g= g.map(plt.scatter,"Sales","Profit",**kws)
g.set(xlim=(0,80),ylim=(0,30))
for ax in g.axes.flat:
    ax.plot((0,80),(0,30),c="orange",ls="--")
g.add_legend()
```
